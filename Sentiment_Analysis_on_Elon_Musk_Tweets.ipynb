**Load data**

```python
import pandas as pd
data = pd.read_csv('/content/Elon_musk.csv', encoding = "ISO-8859-1")
data.head()
```

```python
data.tail()
```

```python
#Load positive and negative words
pos = pd.read_csv('/content/positive-words.txt', encoding = "ISO-8859-1")
neg = pd.read_csv('/content/negative-words.txt', encoding = "ISO-8859-1")
```

```python
pos['sent'] = pd.Series(["pos" for x in range(len(pos.index))])
```

```python
neg['sent'] = pd.Series(["neg" for x in range(len(neg.index))])
```

```python
pos.head()
```

```python
neg.head()
```

```python
#Load stop words
stop = pd.read_csv('/content/stop.txt', encoding = "ISO-8859-1")
stop.head()
```

**Clean the data**

```python
df = data.drop('Unnamed: 0', axis = 1)
df
```

```python
#Removing punctuations
df.replace("[^a-zA-Z]", " ", regex = True, inplace= True)
df
```

```python
pos.replace("[^a-zA-Z]", " ", regex = True, inplace= True)
neg.replace("[^a-zA-Z]", " ", regex = True, inplace= True)
stop.replace("[^a-zA-Z]", " ", regex = True, inplace= True)
```

```python
#Lowering all the letters in data
df['Text'] = df['Text'].str.lower()
df.head()
```

```python
pos['x'] = pos['x'].str.lower()
neg['x'] = neg['x'].str.lower()
stop['x'] = stop['x'].str.lower()
```
